```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
#Import TfIdfVectorizer from scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel
# Compute the Cosine Similarity matrix based on the count_matrix
from sklearn.metrics.pairwise import cosine_similarity
# Import CountVectorizer and create the count matrix
from sklearn.feature_extraction.text import CountVectorizer
from ast import literal_eval
```

```python
tmdb_credits_df=pd.read_csv('InputResources/tmdb_5000_credits.csv')
tmdb_5000_movies_df=pd.read_csv('InputResources/tmdb_5000_movies.csv')
```

```python
#merge the movies and credits
tmdb_credits_df.columns = ['id','tittle','cast','crew']
tmdb_5000_movies_df= tmdb_5000_movies_df.merge(tmdb_credits_df,on='id')
```

```python
####Filtering Based on Vote Average -using IMDB Movie formular#####

mean_votes= tmdb_5000_movies_df['vote_average'].mean()
mean_votes
```

```python
minimum_votes= tmdb_5000_movies_df['vote_count'].quantile(0.9)
minimum_votes
```

```python
movies = tmdb_5000_movies_df.copy().loc[tmdb_5000_movies_df['vote_count'] >= minimum_votes]
movies.shape
```

```python
def weighted_movie_rating(x, m=minimum_votes, C=mean_votes):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
#Use score as the new featurea dand calculate value using weighted_raoing()
movies['score'] = movies.apply(weighted_movie_rating, axis=1)
```

```python
#Sort movies based on score calculated above
movies = movies.sort_values('score', ascending=False)

#What are the top 10 movies?
movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
#CONTENT BASED FILTERING###

tmdb_5000_movies_df['overview'].head(5)
```

```python
#Define a TF-IDF Vectorizer Object. Remove all english stop words such as 'the', 'a'
tfidf = TfidfVectorizer(stop_words='english')

#Replace NaN with an empty string
tmdb_5000_movies_df['overview'] = tmdb_5000_movies_df['overview'].fillna('')

#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(tmdb_5000_movies_df['overview'])

#Output the shape of tfidf_matrix
tfidf_matrix.shape
```

```python
# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#Construct a reverse map of indices and movie titles
indices = pd.Series(tmdb_5000_movies_df.index, index=tmdb_5000_movies_df['title']).drop_duplicates()
```

```python
# Function that takes in movie title as input and outputs most similar movies
def get_movie_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title
    index = indices[title]

    # Get the pairwsie similarity scores of all movies with that movie
    similarity_scores = list(enumerate(cosine_sim[index]))

    # Sort the movies based on the similarity scores
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    similarity_scores = similarity_scores[1:11]

    # Get the movie indices
    movie_indexes = [i[0] for i in similarity_scores]

    # Return the top 10 most similar movies
    return tmdb_5000_movies_df['title'].iloc[movie_indexes]
```

```python
####Credits, Genres and Keywords Based Recommender###

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    tmdb_5000_movies_df[feature] = tmdb_5000_movies_df[feature].apply(literal_eval)
```

```python
# Get the director's name from the crew feature. If director is not listed, return NaN
def get_movie_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
# Returns the list top 3 elements or entire list; whichever is more.
def get_top_three_elements(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        #Check if more than 3 elements exist. If yes, return only first three. If no, return entire list.
        if len(names) > 3:
            names = names[:3]
        return names

    #Return empty list in case of missing/malformed data
    return []
```

```python
# Define new director, cast, genres and keywords features that are in a suitable form.
tmdb_5000_movies_df['director'] = tmdb_5000_movies_df['crew'].apply(get_movie_director)

features = ['cast', 'keywords', 'genres']
for feature in features:
    tmdb_5000_movies_df[feature] = tmdb_5000_movies_df[feature].apply(get_top_three_elements)
```

```python
# Print the new features of the first 3 films
tmdb_5000_movies_df[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
# Convert strings to lower case and strip names of spaces
def format_string_characters(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
# Apply format string function to the features.
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    tmdb_5000_movies_df[feature] = tmdb_5000_movies_df[feature].apply(format_string_characters)
```

```python
def create_metadata(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
tmdb_5000_movies_df['metadata'] = tmdb_5000_movies_df.apply(create_metadata, axis=1)

tmdb_5000_movies_df.head(5)
```

```python
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(tmdb_5000_movies_df['metadata'])
```

```python
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```
